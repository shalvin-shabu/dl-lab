```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
image=cv2.imread("/home/y4ada2/Downloads/images.jpeg",0)
equalized_image=cv2.equalizeHist(image)
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title("original photo")
plt.subplot(1,2,2)
plt.imshow(equalized_image,cmap='gray')
plt.title("equalized photo")

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.hist(image)
plt.title("histogram of original photo")
plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.hist(equalized_image)
plt.title("histogram of equalized photo")
```

```python
kernel=np.ones((5,5),np.uint8)
img=cv2.imread("/home/y4ada2/Downloads/phot.png")
erosion=cv2.erode(img,kernel,iterations=1)
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("original biary image")
plt.figure(figsize=(12,8))
plt.subplot(1,2,2)
plt.imshow(erosion,cmap='gray')
plt.title("eroded image")
```

```python
kernel=np.ones((5,5),np.uint8)
img=cv2.imread("/home/y4ada2/Downloads/phot.png")
dilateion=cv2.dilate(img,kernel,iterations=1)
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("original biary image")
plt.figure(figsize=(12,8))
plt.subplot(1,2,2)
plt.imshow(dilateion,cmap='gray')
plt.title("eroded image")
```
